```python
import numpy as np
import cv2

def resize_and_normalize(data, new_resolution=(256, 256)):
    """
    Resizes and normalizes a dataset of images or masks.

    Parameters:
    - data: A NumPy array of shape (num_samples, height, width, channels) for images
            or (num_samples, height, width) for masks.
    - new_resolution: Tuple specifying the resolution to resize to (default: 256x256).

    Returns:
    - resized_normalized_data: Preprocessed dataset (resized and normalized).
    """
    resized_normalized_data = []

    for sample in data:
        # Resize the sample
        resized_sample = cv2.resize(sample, new_resolution)

        # Normalize the sample
        if np.max(resized_sample) > 1:
            resized_sample = resized_sample / 255.0

        resized_normalized_data.append(resized_sample)

    # Convert list to NumPy array
    resized_normalized_data = np.array(resized_normalized_data, dtype=np.float32)

    return resized_normalized_data

# Load train, validation, and test datasets
train_images = np.load("/content/drive/MyDrive/Robo_final/Numpy/train_images.npy")
train_binary_masks = np.load("/content/drive/MyDrive/Robo_final/Numpy/train_masks.npy")
val_images = np.load("/content/drive/MyDrive/Robo_final/Numpy/val_images.npy")
val_binary_masks = np.load("/content/drive/MyDrive/Robo_final/Numpy/val_masks.npy")
test_images = np.load("/content/drive/MyDrive/Robo_final/Numpy/test_images.npy")
test_binary_masks = np.load("/content/drive/MyDrive/Robo_final/Numpy/test_masks.npy")

# Resize and normalize each dataset
new_resolution = (256, 256)  # Target resolution for resizing

train_images = resize_and_normalize(train_images, new_resolution)
train_binary_masks = resize_and_normalize(train_binary_masks, new_resolution)

val_images = resize_and_normalize(val_images, new_resolution)
val_binary_masks = resize_and_normalize(val_binary_masks, new_resolution)

test_images = resize_and_normalize(test_images, new_resolution)
test_binary_masks = resize_and_normalize(test_binary_masks, new_resolution)

# Save the resized and normalized datasets
np.save('/content/drive/MyDrive/Robo_final/Prepocessed numpy/train_images.npy', train_images)
np.save('/content/drive/MyDrive/Robo_final/Prepocessed numpy/train_binary_masks.npy', train_binary_masks)

np.save('/content/drive/MyDrive/Robo_final/Prepocessed numpy/val_images.npy', val_images)
np.save('/content/drive/MyDrive/Robo_final/Prepocessed numpy/val_binary_masks.npy', val_binary_masks)

np.save('/content/drive/MyDrive/Robo_final/Prepocessed numpy/test_images.npy', test_images)
np.save('/content/drive/MyDrive/Robo_final/Prepocessed numpy/test_binary_masks.npy', test_binary_masks)

print("Resized and normalized train, validation, and test data saved.")
```
